# Azure
You can also access models on [Microsoft Azure](https://azure.microsoft.com/).
You need to sign up at the platform and store API key to your environment as `AZURE_API_KEY`.
You can then access models like shown below.

```python
from bia_bob import bob
```

```python
bob.initialize(endpoint="azure", model="meta-llama-3.1-405b-instruct")
```

```python
%bob Please load blobs.tif and print out its dimensions.
```

```python
import skimage

# Load the image
image = skimage.io.imread("blobs.tif")

# Print out the dimensions of the image
print(image.shape)
```

Alternatively, you can also specify the endpoint yourself:

```python
import os

bob.initialize(endpoint="https://models.inference.ai.azure.com", 
               api_key=os.environ.get('AZURE_API_KEY'),
               model="Phi-3-medium-128k-instruct")
```

```python
%bob Please load blobs.tif and print out its dimensions.
```

```python
from PIL import Image

# Load the TIFF image
with Image.open('blobs.tif') as img:
    # Get the dimensions
    width, height = img.size
    
    # Print the dimensions
    print(f"Image dimensions: {width} x {height}")
```
