```python
from aicsimageio import AICSImage
import stackview

# Load the image file
image_filename = "Image_6_Subset_pos02_t1_35.czi"
aics_image = AICSImage(image_filename)

# Extract the data for all 3 channels at timepoint 9
timepoint = 9
image_data = aics_image.get_image_data("CZYX", T=timepoint)

# Display the image
stackview.switch(image_data)
```

```python
%bob Check in the image metadata for the pixel size of the image in x,y and z dimension
```

```python
# Assume the variable `aics_image` is an AICSImage object loaded with the image data

# Access metadata
metadata = aics_image.physical_pixel_sizes  # Physical pixel sizes is a tuple (z, y, x)

# Extract pixel sizes
pixel_size_z = metadata[0]
pixel_size_y = metadata[1]
pixel_size_x = metadata[2]

# Display the pixel sizes
print(f"Pixel size in x dimension: {pixel_size_x}")
print(f"Pixel size in y dimension: {pixel_size_y}")
print(f"Pixel size in z dimension: {pixel_size_z}")
```
